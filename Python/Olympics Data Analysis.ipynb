```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data

ath = pd.read_csv('C:/Users/B M MORSHED SAYEED/Documents/Data Analysis/Data Analysis Projects/Olympic Data/athlete_events.csv')
reg = pd.read_csv('C:/Users/B M MORSHED SAYEED/Documents/Data Analysis/Data Analysis Projects/Olympic Data/noc_regions.csv')
```

```python
# Now let's take a look at the data

ath.head()
```

```python
reg.head()
```

```python
# join the dataframes usind pandas merge function

ath_df = ath.merge(reg, how = 'left', on = 'NOC')

ath_df.head()
```

```python
# shape of the dataframe

ath_df.shape
```

```python
# Column names consistent

ath_df.rename(columns = {'region':'Region', 'notes':'Notes'}, inplace = True)
```

```python
# all information about the data 

ath_df.info()
```

```python
# description about data (statistical data)

ath_df.describe()
```

```python
# check null values

nan_values = ath_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
ath_df.isnull().sum()
```

```python
# print the column names containing null values or missing values in the from of a list

columns_with_nan = ath_df.columns[nan_columns].tolist()
print(columns_with_nan)
```

```python
# top countries participating

top_10_countries = ath_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# plot for the top 10 countries

plt.figure(figsize= (12,6))
plt.xticks(rotation=20)
plt.title('Overall participation by country')
sns.barplot(x= top_10_countries.index, y = top_10_countries, palette = 'Set2')
```

```python
# age distribution of the participants
# create histogram using matplotlib library

plt.figure(figsize = (12, 6))
plt.title('Age distribution of the Athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(ath_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor = 'white')
```

```python
# Winter olympic sports

winter_sports = ath_df[ath_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer olympic sports

summer_sports = ath_df[ath_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female participants

gender_counts = ath_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female athletes

plt.figure(figsize = (12, 6))
plt.title('Gender distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 180, shadow = True)
```

```python
# Total medals

ath_df.Medal.value_counts()
```

```python
# Total number of female athletes in each olympics

female_participants = ath_df[(ath_df.Sex == 'F') & (ath_df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
# filter data for all female athletes in summer olympics

female_ath = ath_df[(ath_df.Sex == 'F') & (ath_df.Season == 'Summer')]
female_ath.head()
```

```python
# create a count plot usinf seaborn library

plt.figure(figsize= (20, 10))
sns.set(style = "darkgrid")
sns.countplot(x = 'Year', data = female_ath)
plt.title('Women Participation')
```

```python
# participation trends of female athletes

plt.figure(figsize= (20, 10))
plt.title('Plot of the female athletes over time')
part = female_ath.groupby('Year')['Sex'].value_counts()
part.loc[:, 'F'].plot()
```

```python
# Gold medal athletes

gold_medals = ath_df[(ath_df.Medal == 'Gold')]
gold_medals
```

```python
# take only values that are different from NaN (Age isn't NaN)

gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
gold_medals
```

```python
# Gold Medal beyond age = 60

gold_medals['ID'][(gold_medals.Age > 60)].count()
```

```python
# Gold Medal beyond age 60 for which sports

sporting_event = gold_medals['Sport'][(gold_medals.Age > 60)]
sporting_event
```

```python
# plot for sporting_event

plt.figure(figsize = (10, 5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for Athletes over 60 years')
```

```python
# Gold Medals for each countries

totalGold_medals = gold_medals.Region.value_counts().reset_index(name = 'Medals').head(10)
totalGold_medals
```

```python
# Visualize totalGold_medals plot

totalGold_medals = gold_medals.Region.value_counts().reset_index(name = 'Medals').head(6)

g = sns.catplot(x = 'index', y = 'Medals', data = totalGold_medals,
           height = 5, kind = 'bar', palette = 'rocket')
g.despine(left = True)
g.set_xlabels('Top 6 countries')
g.set_ylabels('Number of medals')
plt.title('Gold Medals per Country')
```

```python
# Rio Olympics-2016

max_year = ath_df.Year.max()

team_names = ath_df[(ath_df.Year == max_year) & (ath_df.Medal == 'Gold')].Team.value_counts()
team_names.head(10)
```

```python
# visualize the Rio Olympics data using barplot

sns.barplot(x = team_names.head(15), y = team_names.head(15).index)
plt.xlabel('Countrywise Medals in Rio Olympic-2016')
```

```python
# Filter whole dataset -who have won the medals

not_null_medals = ath_df[(ath_df['Height'].notnull()) & (ath_df['Weight'].notnull()) & (ath_df['Medal'].notnull())]

plt.figure(figsize = (15, 10))
sns.scatterplot(x = 'Height', y = 'Weight', data = not_null_medals, hue = 'Sex')
plt.title('Height vs Weight of Olympics Medalists')
```
